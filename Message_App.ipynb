Message App

```python
import datetime
class User:
  def __init__(self,name):
    self.name = name
    self.timestamp = datetime.datetime.now()
    #store msgs
    self.message = []
    #store blocks
    self.blocked = []

  def send_message(self,receiver,text):
    if self in receiver.blocked:
      print(f"Message not sent - {self.name} is blocked by {receiver.name}!")
      return
    else:
         msg = Message(self,receiver,text)
         receiver.message.append(msg)

  def read_message(self):
    if not self.message:
      print("empty message box")
      return
    for i,msg in enumerate(self.message):
      msg.display_msg()
      msg.mark_as_read()

  def delete_message(self,index):
    if not self.message:
      print(f"empty message box")
      return
    else:
        actual_index = index -1
        del self.message[actual_index]

        print(f'{self.name} deleted the message')

  def block_users(self,receiver):
    self.blocked.append(receiver)
    print(f'{self.name} blocked {receiver.name}')
    return



class Message:
  def __init__(self,sender,receiver,text):
    self.sender = sender
    self.receiver = receiver
    self.text = text
    self.timestamp = datetime.datetime.now()
    self.read = "Unread"

  def mark_as_read(self):
    self.read = "Read"

  def display_msg(self):
    self.mark_as_read()
    print("___New Message!___")
    print(f"FROM: {self.sender.name}")
    print(f"TO: {self.receiver.name}")
    print(f"TEXT: {self.text}")
    print(f"RECEIVED AT: {self.timestamp.strftime('%Y-%m-%d %H:%M')}")
    print(f"STATUS: {self.read}")




user1 = User("Emily")
user2 = User("Lily")
user1.send_message(user2,"Hey!")
user1.send_message(user2,"I SAID HI!")
user2.send_message(user1,"I'm blocking you!!!!")
user2.read_message()
user1.read_message()
user1.delete_message(0)
user2.block_users(user1)
user1.send_message(user2,"I'm Sorry fr")
```
